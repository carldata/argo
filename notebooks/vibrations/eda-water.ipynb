# EDA Water Flow

The following datasets are sampled with the frequency 500Hz

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.figsize'] = 12, 4
```

# Load data and preprocess data

```python
data_frame = pd.read_csv('../../datasets/vibrations/argo-data_water-only.csv')
# Normalize sensor data
data_frame['x_norm'] = (data_frame.x - data_frame.x.mean()) / (data_frame.x.max() - data_frame.x.min())
data_frame['y_norm'] = (data_frame.y - data_frame.y.mean()) / (data_frame.y.max() - data_frame.y.min())
data_frame['z_norm'] = (data_frame.z - data_frame.z.mean()) / (data_frame.z.max() - data_frame.z.min())

data_frame.head()
```

# Plot columns

## Flow

```python
data_frame.flow.plot()
plt.show()
```

## Temperature

```python
data_frame.temp.plot()
plt.show()
```

## Vibration sensor

```python
NFFT = 128
noverlap = NFFT / 4

print('x_norm')
plt.specgram(data_frame.x_norm, Fs=500, NFFT=NFFT, noverlap=noverlap)
plt.show()

print('y_norm')
plt.specgram(data_frame.y_norm, Fs=500, NFFT=NFFT, noverlap=noverlap)
plt.show()

print('z_norm')
plt.specgram(data_frame.z_norm, Fs=500, NFFT=NFFT, noverlap=noverlap)
plt.show()
```

```python
print('x')
plt.psd(data_frame.x, Fs=500)
plt.show()

print('y')
plt.psd(data_frame.y, Fs=500)
plt.show()

print('z')
plt.psd(data_frame.z, Fs=500)
plt.show()
```
