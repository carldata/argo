# EDA Water with the salt Flow

The following datasets are sampled with the frequency 500Hz

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.figsize'] = 12, 4
```

# Load data

```python
data_frame = pd.read_csv('../../datasets/vibrations/argo-data_salt-only.csv')
data_frame.head()
```

# Plot columns

## Flow

```python
data_frame.flow.plot()
plt.show()
```

## Temperature

```python
data_frame.temp.plot()
plt.show()
```

## Vibration sensor

```python
NFFT = 128
noverlap = NFFT / 4

print('x')
plt.specgram(data_frame.x, Fs=500, NFFT=NFFT, noverlap=noverlap)
plt.show()

print('y')
plt.specgram(data_frame.y, Fs=500, NFFT=NFFT, noverlap=noverlap)
plt.show()

print('z')
plt.specgram(data_frame.z, Fs=500, NFFT=NFFT, noverlap=noverlap)
plt.show()
```
